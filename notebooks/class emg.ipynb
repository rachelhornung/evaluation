```python
e.convert_flags(['scale',  'whiten', 'rectify', 'rms', 'arv', 'whiten dispose'])
e.set_order('window size', 'ascend')
e.set_order('latents','ascend')
e.set_order('model', ['LDA', 'QDA', 'LogisticRegression', 'RandomForest', 'GradientBoosting', 'KNN', 'PCA','FA','ICA','VAE','GMM'])
#e.set_order('model', ['LDA', 'QDA', 'LogisticRegression', 'RandomForest', 'GradientBoosting', 'KNN', 'PCA','FA','GMM'])
#e.set_order('flags', ['raw', 'scale', 'whiten', 'scalewhiten', 'scalewhiten dispose',
#                      'rectify', 'scalerectify', 'rectifywhiten', 'rectifywhiten dispose',
#                      'scalerectifywhiten', 'scalerectifywhiten dispose', 
#                      'arv', 'rms', 'scale arv', 'scale rms'])
e.set_order('flags', 'ascend')
e.set_order('file', 'ascend')
```

```python
inds=range(10)
test_strings=[''.join(('test_accuracy', str(i))) for i in inds]
train_strings=[''.join(('train_accuracy', str(i))) for i in inds]
print e.results.shape
e.calc_mean_and_std(attributes=test_strings, output_attributes=['test mean', 'test std'])
e.calc_mean_and_std(attributes=train_strings, output_attributes=['train mean', 'train std'])
print e.results.shape
```
